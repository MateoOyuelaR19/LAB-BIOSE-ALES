### b)	Implemente la multiplicación escalar de a.b ¿Qué se debe realizar para poder multiplicar los dos vectores?

```python
print("Hola Mundo")
#Se procede a crear el par de vectores: 
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
# Debemos definir una función que realice el producto escalar, en esta ocasión se utilizará la función dot de la librería Numpy. 
producto_escalar = np.dot(a,b)
#se imprimirá el resultado
print("El producto escalar de los vectores es:", producto_escalar)
```

### c)	Implemente la multiplicación punto a punto de a.b.

```python
# Debemos definir una función que realice el producto punto, en esta ocasión se utilizará la función multiply de la librería Numpy. 
producto_punto_a_b = np.multiply(a, b)
print("El producto punto a punto de los vectores es:", producto_punto_a_b)
```

###	d) Construya la siguiente matriz:

```python
# Definiremos las filas que formarán la matriz, y posterior a esto, utilizaremos la función array de Numpy para poder operarla como matriz. 
fila1 = [2, -1, -3]
fila2 = [4, 1.5, -2.5]
fila3 = [7.3, -0.9, 0.2]
A= np.array([fila1, fila2, fila3])
print("La matriz A, es:")
print(A)
```

### f)	Consulte la función de los siguientes comandos de la librería numpy: ones, round, ceil, floor, implemente cada función para ejemplificar su uso.

```python
#La Función Ones de Numpy, crea una matriz de unos, en este caso la emplearemos en la creación de una matriz 2x3 de unos:
arr_ones = np.ones((2, 3))
print("E) Array de unos:")
print(arr_ones)

#La función rounded se utiliza para redondear los elementos de un arreglo a la cantidad de decimales especificada, en este caso crearemos un array con números decimales y usaremos la función para redondear los decimales en 2. 
arr_decimal = np.array([3.14159, 2.71828, 1.41421])
arr_rounded = np.round(arr_decimal, decimals=2)
print("Array redondeado:")
print(arr_rounded)

#Función ceil de Numpy se utiliza para calcular el techo de cada elemento en un arreglo, es decir, redondea cada elemento hacia arriba al entero más cercano utilizando el array de decimales creado anteriormente, la función floor, es el contrario de la ceil. Llamaremos a la función:
# Crear un array con números decimales
# Calcular el techo de cada elemento del array
arr_ceil = np.ceil(arr_decimal)
# Calcular el piso de cada elemento del array
arr_floor = np.floor(arr_decimal)
print("G) Techo de cada elemento del array:")
print(arr_ceil)
print("H) Piso de cada elemento del array:")
print(arr_floor)
```

### j) Construya la función  y[n]=sin⁡(π*0.12n) en el intervalo 0 <= n <= 100.

```python
# Definiremos el rango de valores de n segun lo indicado:
n_values = np.arange(0, 101)

# Calcular los valores de la función y[n] = sin(π * 0.12 * n) usando numpy
y_values = np.sin(np.pi * 0.12 * n_values)

# Graficar la función, aqui por primera vez utilizaremos las funciones de la librería Plt:
plt.plot(n_values, y_values)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Función y[n] = sin(π * 0.12 * n)')
plt.grid(True)
plt.show()
```

### k) Construya otra senoidal y2[n]=cos⁡(2π*0.03n) 

```python
# Calcular los valores de la función y2[n] = cos(2π * 0.03 * n), seguimos tomando los valores de 0 a 100 creados en la seccion anterior:
y2_values = np.cos(2 * np.pi * 0.03 * n_values)

# Graficar la función
plt.plot(n_values, y2_values)
plt.xlabel('n')
plt.ylabel('y2[n]')
plt.title('Función y2[n] = cos(2π * 0.03 * n)')
plt.grid(True)
plt.show()
```

### l)Genere una tercera señal que sea la suma de estas dos señales, es decir 
### s[n]=y[n]+y2[n] y una cuarta, que sea el producto de estas, t[n]=y[n].y2[n].


```python
#PARA ESTA UTILIZAREMOS LOS VALORES OBTENIDOS EN LOS APARTADOS ANTERIORES DE LAS FUNCIONES REALIZADAS
# Calcularemos la cuarta señal t[n] = y[n] * y2[n]
# Calcular la tercera señal s[n] = y[n] + y2[n]
s_values = y_values + y2_values

# Calcular la cuarta señal t[n] = y[n] * y2[n]
t_values = y_values * y2_values

# Graficar las señales
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(n_values, s_values, color='blue')
plt.xlabel('n')
plt.ylabel('s[n]')
plt.title('Tercera señal s[n] = y[n] + y2[n]')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(n_values, t_values, color='red')
plt.xlabel('n')
plt.ylabel('t[n]')
plt.title('Cuarta señal t[n] = y[n] * y2[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### m) Graficar en la misma figura las señales y[n] y y2[n], utilice leyenda y colores diferentes para cada señal. Asigne título a cada eje.

```python
# Graficar las señales en la misma figura
plt.plot(n_values, y_values, label='y[n]', color='blue')
plt.plot(n_values, y2_values, label='y2[n]', color='red')

# Agregar leyenda, againn
plt.legend()

# Título de los ejes, againnnnnnn
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señales y[n] y y2[n]')

# Mostrar agaiiiiiiiiinnn
plt.grid(True)

# Mostrar la figura
plt.show()
```

### n)	Graficar en la misma figura las señales s[n] y t[n], utilice leyenda y colores diferentes para cada señal. Asigne título a cada eje.

```python
# Calcularemos los valores de s[n] y t[n]
s_values = np.sin(np.pi * 0.12 * n_values) + np.cos(2 * np.pi * 0.03 * n_values)
t_values = np.sin(np.pi * 0.12 * n_values) * np.cos(2 * np.pi * 0.03 * n_values)

# Graficaremos las señales en la misma figuraaa
plt.plot(n_values, s_values, label='s[n]', color='blue')
plt.plot(n_values, t_values, label='t[n]', color='red')

# Agregaermos la leyenda
plt.legend()

# Título de los ejess
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.title('Señales s[n] y t[n]')

# Mostrar la rejilla againn
plt.grid(True)

# Mostrar la figura againnn
plt.show()
print("Adiós Mundo")
```
